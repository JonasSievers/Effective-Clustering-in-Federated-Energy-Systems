```python
#Imports

#Data handling
import pandas as pd
#Create data arrays
import numpy as np
```

```python
prosumption = pd.read_csv("../Final_Prosumption_dataset.csv")
prosumption.columns
```

```python
non_user_columns = [col for col in prosumption if 'User' not in col]

# Print non-user columns
print(non_user_columns)
```

```python
unrelated_features = ['Date', 'temp', 'rhum', 'wspd', 'PC1', 'hour sin', 'hour cos']

# Create a list of user columns to keep, from 'User1' to 'User70'
user_columns_to_keep = [f'User{i}' for i in range(1, 71)]
user_lag_columns_to_keep = [f'User{i}_lag_24hrs' for i in range(1, 71)]
# Combine the lists of columns to keep
columns_to_keep = unrelated_features + user_columns_to_keep + user_lag_columns_to_keep

# Filter the dataframe to keep only the selected columns
filtered_df = prosumption[columns_to_keep]

# Display the filtered dataframe
filtered_df.head()
```

```python
filtered_df.to_csv('Final_Prosumption_dataset.csv', index=False)
```
